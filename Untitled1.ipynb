```r
require('data.table')
require('stringr')
require('ggplot2')
require('plyr')
require('RColorBrewer')

df = fread('EDMP Survey.csv')


colnames(df) <- c('Timestamp', 'DAW', 'DAW_Switch', 'DAW_Which', 'Favorite_Synth', 'Favorite_Effect', 'Hardware_synth',
                 'Modular', 'Monitors', 'Monitor_Brand', 'Headphones', 'Headphones_Brand', 'Audio_Interface', 'Audio_Brand'
                 ,'Audio_Inputs', 'Audio_Outputs', 'Audio_Protocol', 'Native_Language', 'Age', 'Channel_Added', 'Rule_Added',
                 'Q&A', 'Events', 'Misc')
```

```r
df.all_daws <- df[,.(unlist(str_split(DAW, ";")), DAW_Which),by=DAW_Switch]
colnames(df.all_daws) <- c('DAW_Switch', 'DAW', 'DAW_Which')

row_numbers<-grep(paste(c('Hardware','HARDWARE','Masch','Korg'),collapse="|"), df.all_daws$DAW,)
for (i in row_numbers) df.all_daws$DAW[i] <- "Hardware"

row_numbers<-grep(paste(c('Aud','Sony','Minecraft','Digital'),collapse="|"), df.all_daws$DAW,)
for (i in row_numbers) df.all_daws$DAW[i] <- "Other"

row_numbers<-grep(paste(c('Aud','Sony','Minecraft','Digital'),collapse="|"), df.all_daws$DAW_Which,)
for (i in row_numbers) df.all_daws$DAW_Which[i] <- "Other"

row_numbers<-grep(paste(c('Renoise'),collapse="|"), df.all_daws$DAW,)
for (i in row_numbers) df.all_daws$DAW[i] <- "Renoise"

df.all_daws.count <- count(df.all_daws)[-1,]
setDT(df.all_daws.count)
df.all_daws.count.yes <- subset(df.all_daws.count, DAW_Switch=='Yes')
df.all_daws.count.no <- subset(df.all_daws.count, DAW_Switch=='No')
```

```r
reaper = '#6c176d'
cubase = '#2c6889'
bitwig = '#9e2712'
protools = '#a870af'
reason  = '#960196'
renoise = '#35d127'
logicGB = '#481470'
mixcraft = '#0c5116'
magix = '#6a756c'
s1 = '#eadb33'
ableton = '#db6811'
flstudio = '#0c994b'
other = '#ff7d05'
hardware = '#d60629'
color_palette = c(ableton,bitwig,cubase,flstudio,hardware,logicGB,magix,mixcraft,other,protools,reaper,reason,renoise,s1)
```

```r
ggplot(df.all_daws.count.no, aes(x=DAW,y=freq, fill = DAW),) + 
    geom_bar(width = .5, stat = 'identity',color='black')+coord_flip() +
    geom_bar(data = df.all_daws.count.yes, stat='identity', 
             aes(fill=df.all_daws.count.yes$DAW_Which), color='black', alpha= 1) + 
    theme_classic() + theme(legend.position  = 'top') +
    scale_fill_manual(values = color_palette)
```

```r
trim <- function (x) gsub("^\\s+|\\s+$", "", x)
first.word <- function(my.string){
    unlist(strsplit(my.string, " "))[-1]
}
Monitors$Model_Name <- sapply(Monitors[,Monitor_Brand], first.word)

temp <- df[,.(Monitors,Monitor_Brand)][Monitors == 'Yes']
Monitors$Monitor_Brand <-temp[, .(unlist(strsplit(Monitor_Brand, ",")))]
Monitors$Monitor_Brand <- Monitors[,(tolower(Monitor_Brand))]
Monitors$Monitor_Brand <- Monitors[,(trim(Monitor_Brand))]
Monitors$Brand_Name <- vapply(str_split(Monitors$Monitor_Brand," "), `[`, 1, FUN.VALUE=character(1))
Monitors$Model_Name





row_numbers<-grep(paste(c('rok','krk'),collapse="|"), Monitors$Monitor_Brand,)
for (i in row_numbers) Monitors$Brand_Name[i] = "krk"

#Monitors$Model_Name <- gsub(paste(c('rocket', 'roxit'),collapse = "|"),'rokit',Monitors$Model_Name)
Monitors$Model_Name <- gsub("'s",'',Monitors$Model_Name)
Monitors$Model_Name <- gsub("reveal",'',Monitors$Model_Name)


row_numbers<-grep(paste(c('yamah'),collapse="|"), Monitors$Monitor_Brand,)
for (i in row_numbers) Monitors$Brand_Name[i] <- "yamaha"

row_numbers<-grep(paste(c('m-audio'),collapse="|"), Monitors$Monitor_Brand,)
for (i in row_numbers) Monitors$Brand_Name[i] <- "m-audio"

row_numbers<-grep(paste(c('mind','thx',"don't", "not sure", "-", "^d$", "idk", "shitty", "old", "^ "),collapse="|"), Monitors$Monitor_Brand,)
for (i in row_numbers) {Monitors$Brand_Name[i] <- NA
                        Monitors$Model_Name[i] <- NA}

row_numbers<-grep(paste(c('cheap','self','walmart'),collapse="|"), Monitors$Monitor_Brand,)
for (i in row_numbers) {Monitors$Brand_Name[i] <- "other"
                        Monitors$Model_Name[i] <- "other"}


row_numbers<-grep(paste(c('305'),collapse="|"), Monitors$Monitor_Brand,)
for (i in row_numbers) Monitors$Model_Name[i] <- "lsr305"
```
